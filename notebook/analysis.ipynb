```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_videos = df.sort_values(by='views', ascending=False).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x='views',
    y='title',
    data=top_videos
)

plt.title("Top 10 Most Viewed Videos")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='views',
    y='likes',
    data=df
)

plt.title("Views vs Likes")

plt.show()
```

```python
top_channels = df['channel_title'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_channels.values,
    y=top_channels.index
)

plt.title("Top Channels")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df[['views', 'likes', 'dislikes', 'comment_count']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
top_categories = df['category_id'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_categories.index,
    y=top_categories.values
)

plt.title("Most Popular Categories")

plt.xlabel("Category ID")
plt.ylabel("Number of Videos")

plt.show()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])
```

```python
df['publish_year'] = df['publish_time'].dt.year

yearly_uploads = df['publish_year'].value_counts().sort_index()

plt.figure(figsize=(8,5))

sns.lineplot(
    x=yearly_uploads.index,
    y=yearly_uploads.values
)

plt.title("Video Upload Trend by Year")

plt.xlabel("Year")
plt.ylabel("Uploads")

plt.show()
```

1. Videos with higher likes usually have higher views.
2. Some channels upload trending content frequently.
3. Entertainment categories dominate trending videos.
4. Engagement strongly correlates with views.
